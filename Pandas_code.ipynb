# Remaning Axis Indexs


```python
data=pd.DataFrame(np.arange(16).reshape((4,4)),index=["a","b","c","d"],columns=["two","three","four","Five"])
data
```

```python
def transform(x):
  return x[:4].upper()
```

```python
data.index.map(transform)
```

```python
data.index=data.index.map(transform)
```

```python
data
```

```python
data.rename(index=str.title,columns=str.upper)
```

```python
data.rename(index={"A":"January"},
            columns={"four":"Zero"})
```
