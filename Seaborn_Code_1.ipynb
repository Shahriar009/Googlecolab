```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
#Loading a default dataset 'titanic'
titanic=sns.load_dataset('titanic')
titanic
```

```python
from re import X
#creating a histogram plot of the  'fare'  variable

sns.histplot(data=titanic,x="fare")

```

```python
#creating a histogram plot of the  'age' variable
sns.histplot(data=titanic,x="age")
```

```python
#creating a histogram plot using displot
sns.displot(data=titanic,x="age",kind="hist")
```

```python
#  (Hue)Define which color represent which things
sns.displot(data=titanic,x='age',kind='hist',bins=4,hue='sex',element='step')
```

```python
sns.displot(data=titanic,x="age",kind="hist",bins=8)
```

```python
sns.displot(data=titanic,x='age',kind='hist',hue='class',element='step')
```

```python
sns.displot(data=titanic,x="age",kind="hist",hue="alive",element='step')
```

```python
sns.kdeplot(data=titanic,x='age')
```

```python
sns.displot(data=titanic,x='age',kind='kde',hue='sex',fill=True,height=4,aspect=2)
```

```python
sns.displot(data=titanic,x='age',kind='kde',hue='class',fill=True,height=4,aspect=2)
```

```python
sns.displot(data=titanic,x='age',kind='kde',hue='alive',fill=True,height=2,aspect=2)
```

```python
#Creating a violin plot using violinplot

sns.violinplot(data=titanic,x="age",y="alive")
#set the title and label

plt.title("alive -violin plot")

plt.xlabel("Age of person")

plt.ylabel("alive")

plt.show()
```
