```python
colors=['#06d6a0','#2ec4b6','#ffbf69','#cbf3f0']
```

```python
data={
    'Position of Relagions':[1,4,2,3],
    'Name of relagions':['Islam','Hindu','Buddho','Kristhan']
}
df=pd.DataFrame(data)
df
```

```python
fig,ax=plt.subplots()

bars=ax.barh(df['Name of relagions'],df['Position of Relagions'],color=colors)

for i, bar in enumerate(bars):
  ax.text(bar.get_width() +0.3, bar.get_height() / 2, df['Position of Relagions'].iloc[i],ha='left',va='top',fontsize=10)
ax.set_xlabel('Position of Relagions')
ax.set_title('Total Number of persons in Different Relagions')
ax.set_xlim([0,5])
ax.set_ylim([0,4])
plt.show()
```

```python
# Renaming column

df.rename(columns={'Products_Model':'Model',
                   'Products_link':'Link',
                   'Products_details':'Details'},inplace=True)
```

Creating a new DataFrame object consisting of two columns

```python
df_g=df[['Model','Details']]
```

```python
df_g
```

#Group by on Model column,sum aggreation


```python
#Reseting index of Model to column
df_g=df_g.groupby(by='Model').sum()
```

```python
df_g
```

```python
df_g.reset_index(inplace=True)
```

```python
df_g
```

```python
#Create Figure
```

```python
#Plot area
#Creating Figure Object
```

```python
#Axes
```

```python
figure=plt.figure(figsize=(4,4))
ax=fig.add_axes([0,0,1,1])
ax.set_xlabel('Details')
ax.set_ylabel('Model')
ax.set_title('Device spaces top products')
#Bar plot
plt.bar(x=df_g['Details'],height=df_g['Model'])
plt.show()
```
