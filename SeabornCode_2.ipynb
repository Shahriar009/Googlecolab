```python
flights=sns.load_dataset('flights')
flights
```

```python
#Pivot is the data to create a matrix format

flights_pivot=flights.pivot("year","month","passengers")

#creating heatmap

sns.heatmap(data=flights_pivot, cmap="BuPu")

plt.title("Passenger Tarfic Heatmap")

plt.xlabel("Months")
plt.ylabel("Year")

plt.show()
```

```python
#Pivot is the data to create a matrix format

flights_pivot=flights.pivot("year","month","passengers")

#creating heatmap
fig, ax = plt.subplots(figsize=(12, 8)) #Fixed the size of matrix


sns.heatmap(data=flights_pivot, cmap="BuPu",annot=True,ax=ax) # to show values(annot=True) and for size(ax=qx)

plt.title("Passenger Tarfic Heatmap")

plt.xlabel("Months")
plt.ylabel("Year")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as pl
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
#Creating a Facegrid with rows based on survive and columns based on sex
grid=sns.FacetGrid(data=titanic,row="survived",col="sex")

#Maping a scatterplot onto each subset of data
grid.map(sns.scatterplot,"fare","pclass")
```

```python
grid=sns.FacetGrid(data=titanic,row="survived",col="sex")


grid.map(sns.barplot,"fare","pclass")

plt.show
```

```python
planets=sns.load_dataset('planets')
planets
```

```python
#(Facing problem)

#Creating a Facegrid with rows based on survive and columns based on sex
grid=sns.FacetGrid(data=planets,row="method",col="year")

#Maping a scatterplot onto each subset of data

grid.map(sns.scatterplot,"orbital_period","distance")

plt.show()
```
