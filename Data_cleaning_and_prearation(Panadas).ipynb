Discretization and Binning

```python
import numpy as np
import pandas as pd
```

```python
 ages = [11,22,33,44,55,66,77,88,99,17,34,68]
```

```python
bins=[15,22,30,50,110
      ]
```

```python
age_catagories=pd.cut(ages,bins)
```

```python
age_catagories
```

```python
age_catagories.codes
```

```python
information={"ages":[11,22,33,44,55,66,77,88,99,17,34,68],
             "range":[15,22,30,50,110,12,2,4,5,6,7,8]}
```

```python
df=pd.DataFrame(information)
df
```

```python
age_catagories.categories
```

```python
age_catagories.categories[0
                          ]
```

```python
pd.value_counts(age_catagories)
```

```python
group_names=["Junior","Adult","MiddleAged", "Seniour"]
```

```python
label_cut=pd.cut(ages,bins,labels=group_names)
```

```python
pd.value_counts(label_cut)
```

```python
data=np.random.uniform(size=25)
data
```

```python
pd.cut(data,6,precision=3)
```

```python
data=np.random.standard_normal(1000)
```

```python
data
```

```python
quartiles=pd.qcut(data,5,precision=4)
quartiles
```

```python
pd.value_counts(quartiles)
```
