Combining and Merging Datasets

```python
dataframe1=pd.DataFrame({"Key":["A","A","B","C","B","B","A","A","C"],
                         "data":pd.Series(range(9),dtype="Int64")})
dataframe1
```

```python
dataframe2=pd.DataFrame({"Key":["A","B","D"],
                         "data2":pd.Series(range(3),dtype=
                                           "Int64")})
dataframe2
```

```python
pd.merge(dataframe1,dataframe2)
```

```python
dataframe3=pd.DataFrame({"ikey":["A","A","B","C","B","B","A","A","C"],
                         "data1":pd.Series(range(9),dtype="Int64")})
dataframe3
```

```python
dataframe4=pd.DataFrame({"ktype":["A","B","D"],
                         "data2":pd.Series(range(3),dtype="Int64")})
dataframe4
```

```python

pd.merge(dataframe3,dataframe4)
```

```python
 pd.merge(dataframe3,dataframe4,left_on="ikey", right_on="ktype")
```

```python
 pd.merge(dataframe3, dataframe4, how="outer")
```

```python
 pd.merge(dataframe3,dataframe4,on="ikey",how="left")
```

```python
dataframe5=pd.DataFrame({"ttype":["A","B","D"],
                         "data2":pd.Series(range(3),dtype="Int64")})
dataframe5
```

```python
dataframe6=pd.DataFrame({"ykey":["A","A","B","C","B","B","A","A","C"],
                         "data1":pd.Series(range(9),dtype="Int64")})
dataframe6
```

```python
pd.merge(dataframe5,dataframe6)
```

```python
pd.merge(dataframe5,dataframe6,how="inner")
```
